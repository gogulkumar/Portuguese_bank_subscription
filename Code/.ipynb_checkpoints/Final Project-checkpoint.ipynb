# Data Preprocessing

```python
numerical_columns = [c for c in trainData.columns if trainData[c].dtype == 'int64' or trainData[c].dtype == 'float64']
categorical_columns = [c for c in trainData.columns if trainData[c].dtype == 'object' and c != 'has_subscribed']
```

```python
def create_encoding_map(data, encodingMap = {}, excludeColumns = set()):
    for column in data.columns:
        if data[column].dtypes == 'object' and column not in excludeColumns:
            if column not in encodingMap:
                encodingMap[column] = {
                    "label": {value: index for index, value in enumerate(data[column].unique())},
                    "value": {index: value for index, value in enumerate(data[column].unique())}
                }
            else:
                nextValue = max(encodingMap[column]['value']) + 1
                for eachLabel in set(encodingMap[column]['label'].keys()) ^ set(data[column].unique()):
                    encodingMap[column]["label"][eachLabel] = nextValue
                    encodingMap[column]["value"][nextValue] = eachValue

    return encodingMap
```

```python
def encode_labels(data, encodingMap):
    for columnName in encodingMap:
        data[columnName] = data[columnName].map(encodingMap[columnName]['label'])

    return data
```

```python
def onehot_encode(data, columnName, encodingMap, includeExistingColumn = False):
    currentColumnData = None
    if includeExistingColumn:
        currentColumnData = data[columnName].copy(deep = True)
    data = pd.get_dummies(
        data=data,
        columns=[columnName]
    )
    insertedColumns = set()
    
    for column in data.columns:
        if columnName in column and column.split("_")[-1].isnumeric():
            editedColumnName = column.split("_")
            editedColumnName[-1] = f"{encodingMap[columnName]['value'][int(editedColumnName[-1])]}"
            insertedColumns.add(editedColumnName[-1])
            data = data.rename(columns={column: "_".join(editedColumnName)})
            
    for column in encodingMap[columnName]['label']:
        if column not in insertedColumns:
            insertColumnName = f"{columnName}_{column}"
            data[insertColumnName] = 0
            
    if includeExistingColumn:
        data[columnName] = currentColumnData
        
    return data
```

```python
encodingMap = create_encoding_map(trainData, encodingMap = {})
trainData = encode_labels(trainData, encodingMap)
for columnName in categorical_columns:
    trainData = onehot_encode(trainData, columnName, encodingMap, includeExistingColumn = True)
    
target = trainData.pop('has_subscribed')
trainData['has_subscribed'] = target
```

```python
encodingMap = create_encoding_map(trainData, encodingMap = encodingMap)
testData = encode_labels(testData, encodingMap)
for columnName in categorical_columns:
    testData = onehot_encode(testData, columnName, encodingMap)

target = testData.pop('has_subscribed')
testData['has_subscribed'] = target
```

## Removing Outliers

```python
def remove_outliers(data, columnName):
    Q1 = np.percentile(data[columnName], 25, interpolation = 'midpoint')
    Q3 = np.percentile(data[columnName], 75, interpolation = 'midpoint')
    IQR = Q3 - Q1
    
    data = data[(data[columnName] <= (Q3 + 1.5 * IQR)) & (data[columnName] >= (Q1 - 1.5 * IQR))]
    data.reset_index(drop = True, inplace = True)
    
    return data
```

```python
outlier_columns = ['age']
for column in outlier_columns:
    data = remove_outliers(data, column)
    
# specific for campaign
data = data[(data['campaign'] <= 20)]
```

## Scaling the data for numerical columns

```python
def scale_data(data, numerical_columns):
    col_names = numerical_columns
    features = data[col_names]
    scaler = StandardScaler().fit(features.values)
    features = scaler.transform(features.values)
    data[col_names] = features
    return data
```

```python
'''function to evaluate the performance of the model'''
def evaluate(y, y_hat):
    '''
    Parameters:
        y: the actual resultant value 
        y_hat: the predicted value obtained from the respective model
    '''
    y = (y == 1)
    y_hat = (y_hat == 1)
    accuracy = ( y == y_hat).sum() / y.size
    precision = (y & y_hat).sum() / y_hat.sum()
    recall = (y & y_hat).sum() / y.sum()
    f1score = 2 * (precision * recall) / (precision + recall)

    return accuracy, precision, recall, f1score
```

# 1.2) Custom Logistic Regression Model

```python
class LogisticRegression:
    """
    Parameters:
        X: features
        y: target variable
        learningRate: 0.00001
        tolerance: 0.0001
    """
    def __init__(self,X, y, X_test_unseen, y_test_unseen, learningRate, tolerance, maxIteration=100000,
                 smote_ind=False, PCA_ind = False):
        self.X=X
        self.y=y
        self.X_test_unseen = X_test_unseen
        self.y_test_unseen = y_test_unseen
        self.learningRate=learningRate
        self.tolerance=tolerance
        self.maxIteration=maxIteration
        self.smote_ind = smote_ind
        self.PCA_ind = PCA_ind
        
    '''function to scale the train data'''
    def normalize(self,X):
        mean=np.mean(X)
        std=np.std(X)
        X_norm=(X-mean)/std
        X_norm=self.addX0(X_norm)
        
        return X_norm,mean,std
    
    '''scale the test data'''
    def normalizeTestData(self,X_test,train_mean,train_std):
        X_norm=(X_test-train_mean)/train_std
        X_norm=self.addX0(X_norm)
        
        return X_norm
        
    '''add bias to the dataset'''
    def addX0(self,X):
        return np.column_stack([np.ones([X.shape[0],1]),X])
    
    '''calculating the probability of belonging to a particular class'''
    def sigmoid(self,z):
        sig=1/(1+np.exp(-z))
        return sig
    
    '''returning the cost of the model'''
    def costFunction(self,X,y):
        pred_= np.log(np.ones(X.shape[0])+np.exp(X.dot(self.w))) - X.dot(self.w)*y #negative log-likelihood
        cost = pred_.sum()
        
        return cost
    
    '''calculating the gradient'''
    def gradient(self,X,y):
        sig=self.sigmoid(X.dot(self.w))
        grad=(sig-y).dot(X)
        return grad
    
    '''gradient descent algorithm'''
    def gradientDescent (self,X,y):
        costSequence=[]
        lastCost=float('inf')
        
        for i in tqdm(range(self.maxIteration)):
            self.w=self.w-self.learningRate*self.gradient(X,y)
            currentCost = self.costFunction(X,y)
            diff = lastCost-currentCost
            lastCost=currentCost
            costSequence.append(currentCost)
            
            if abs(diff) < self.tolerance:
                print("The Model Stopped - No Further Improvement")
                break
                
        self.plotCost(costSequence)
        print(diff)
        return
    
    '''plot the cost of the model'''    
    def plotCost(self,costSequence):
        s=np.array(costSequence)
        t=np.arange(s.size)
        
        fig,ax=plt.subplots()
        ax.plot(t,s)
        
        ax.set(xlabel='iterations',ylabel='cost',title='cost trend')
        ax.grid()
        
        plt.legend(bbox_to_anchor=(1.05,1),shadow=True)
        plt.show()
       
    '''predicted classes'''
    def predict(self,X):
        sig = self.sigmoid(X.dot(self.w))
        return np.around(sig)
    
    '''performance of the model'''
    def evaluate(self,y,y_hat):
        y = (y==1)
        y_hat = (y_hat == 1)

        accuracy = ( y == y_hat).sum()/ y.size
        precision = (y & y_hat).sum() / y_hat.sum()
        recall = (y & y_hat).sum() / y.sum()
        f1score = 2 * (precision * recall) / (precision + recall)
        
        return accuracy,precision,recall,f1score
    
    '''run the model'''
    def fit(self):
        self.X_train,self.X_test,self.y_train,self.y_test= train_test_split(self.X,
                                                                            self.y,
                                                                            stratify=self.y,
                                                                            test_size=0.20,
                                                                            random_state=0)
        if self.smote_ind == True:
            smote = SMOTETomek(random_state=0)
            self.X_train,self.y_train = smote.fit_resample(self.X_train,self.y_train)
            
        self.X_train,self.mean,self.std=self.normalize(self.X_train)
        self.X_test=self.normalizeTestData(self.X_test,self.mean,self.std)
        
        
        self.w = np.ones(self.X_train.shape[1],dtype=np.float64)*0
        self.gradientDescent(self.X_train,self.y_train)
        
        y_hat_train = self.predict(self.X_train)
        self.accuracy,self.precision,self.recall,self.f1score=self.evaluate(self.y_train,y_hat_train)
        print("=" * 50)
        print('Metrics for y_train:')
        print("=" * 50)
        metrics_labels = ['Accuracy', 'Precision', 'Recall', 'F1 score']
        metrics_values = [[self.accuracy, self.precision, self.recall, self.f1score]]
        print(pd.DataFrame(metrics_values, columns = metrics_labels, index = ['value']))
        print(f"\nConfusion Matrix:\n{pd.DataFrame(confusion_matrix(self.y_train, y_hat_train))}\n")

        y_hat_test = self.predict(self.X_test)
        self.accuracy,self.precision,self.recall,self.f1score=self.evaluate(self.y_test,y_hat_test)
        print("=" * 50)
        print('Metrics for y_test:')
        print("=" * 50)
        metrics_labels = ['Accuracy', 'Precision', 'Recall', 'F1 score']
        metrics_values = [[self.accuracy, self.precision, self.recall, self.f1score]]
        print(pd.DataFrame(metrics_values, columns = metrics_labels, index = ['value']))
        print(f"\nConfusion Matrix:\n{pd.DataFrame(confusion_matrix(self.y_test, y_hat_test))}\n")
        if self.PCA_ind==False:
            self.unseen_test(self.X_test_unseen,self.y_test_unseen)
        
        
    def unseen_test(self,X_test,y_test):
        X_test=self.normalizeTestData(X_test,self.mean,self.std)
        y_hat_test = self.predict(X_test)
        accuracy,precision,recall,f1score=self.evaluate(y_test,y_hat_test)
        metrics_labels = ['Accuracy', 'Precision', 'Recall', 'F1 score']
        metrics_values = [[accuracy, precision, recall, f1score]]
        print("=" * 50)
        print('Metrics for UNSEEN y_test:')
        print("=" * 50)
        print(pd.DataFrame(metrics_values, columns = metrics_labels, index = ['value']))
        print(f"\nConfusion Matrix:\n{pd.DataFrame(confusion_matrix(y_test, y_hat_test))}\n")       
        
```

## 2.2) Buliding PCA model

```python
trainData_X, testData_X = copy_trainData_X.copy(deep = True), unseen_testData_X.copy(deep = True)
```

```python
pca = PCA()
pca.fit(trainData_X)

var= pca.explained_variance_ratio_
var1=np.cumsum(np.round(pca.explained_variance_ratio_, decimals=4)*100)
plt.plot(var1)
```

```python
''' This function counts the actual number of features that impacts the dataset'''
def optimal_feature_count():
    percentage_addition_of_features = var[0]
    for i in range(1,len(var)):
        if percentage_addition_of_features+var[i]-percentage_addition_of_features!=0:
            percentage_addition_of_features = percentage_addition_of_features + var[i]
        else:
            break
    return i+1
optimal_feature_count()   
```

```python
'''
pca - Princpal Component Analysis Variable
components - transforming the projected pca to actual values
X_temp - Dataset of the transformed variable
'''
pca = PCA(n_components=optimal_feature_count())
pcs = pca.fit(trainData_X)
components = pcs.transform(trainData_X)
#modifying Train data
X_temp = pd.DataFrame(components)
X_temp = X_temp.reset_index(drop = True)
y_train = trainData_y.reset_index(drop=True)
X_temp = pd.concat([X_temp, y_train],axis=1)
```

## 2.4) Using the custom SVM model

```python
"""
X_temp_0 - dataframe that holds the value of 0 
X_temp_1 - dataframe that holds the value of 1
X_temp   - concated dataframe of X_temp_0, X_temp_1
"""
X_temp[X_temp['has_subscribed']==0] = -1
X_temp_test[X_temp_test['has_subscribed']==0] = -1
X_temp_0 = X_temp[X_temp['has_subscribed']==-1]
X_temp_1 = X_temp[X_temp['has_subscribed']==1]
X_temp_0 = X_temp_0.sample(n=1000,random_state =0 )
X_temp_1 = X_temp_1.sample(n=1000,random_state =0)
X_temp = pd.concat([X_temp_0,X_temp_1])
```

```python
"""
X_temp_x - holds the independent features of the PCA transformed(X_temp)
X_temp_x - holds the dependent features of the PCA transformed(X_temp)
"""
X_temp_x = X_temp.iloc[:,:-1]
X_temp_y = X_temp.iloc[:,-1]
```

```python
class SoftMarginSVM:
    """
    Parameters:
        X: features
        y: target variable
        w: weights(parameters which are learnt)
        C: regularization term
    """
    def __init__(self, X, y, X_test_unseen, y_test_unseen, C):
        self.alpha = None
        self.w = None
        self.supportVectors = None
        self.X = X
        self.y = y
        self.C = C
        self.X_test_unseen = X_test_unseen
        self.y_test_unseen = y_test_unseen
        
    '''fitting the model'''
    def fit(self, X, y):
        N = len(y)

        # Gram matrix of (X,y)
        Xy = X*y[:,np.newaxis] 

        GramXy = np.matmul(Xy,Xy.T) 
        
        '''the dual form of SVM'''
        def Ld0(G, alpha): 
            obj_fn = alpha.sum() - 0.5*alpha.dot(alpha.dot(G))
            return obj_fn
        

        '''
        Derivative of Lagrangian Function
        '''
        def partialDerivationLd0(G, alpha):
            par_der = np.ones_like(alpha) - alpha.dot(G)
            return par_der

        alpha = np.ones(N)

        A = np.vstack((-np.eye(N), np.eye(N)))
        b = np.concatenate((np.zeros(N), self.C * np.ones(N)))

        constraints = ({'type': 'eq', 'fun':lambda a: np.dot(a,y), 'jac': lambda a: y },
                    {'type': 'ineq', 'fun':lambda a: b - np.dot(A, a), 'jac': lambda a: -A}) 

        optRes = optimize.minimize(fun = lambda a: -Ld0(GramXy, a),
                                x0 = alpha,
                                method = 'SLSQP',
                                jac = lambda a: - partialDerivationLd0(GramXy, a),
                                constraints = constraints)
        self.alpha = optRes.x
        self.w = np.sum(( self.alpha[:, np.newaxis] * Xy), axis = 0)
        epsilon = 1e-3
        self.supportVectors = X[ self.alpha > epsilon]
        self.supportLabels = y[self.alpha > epsilon]
        self.supportVectors = np.array(self.supportVectors)
        self.supportLabels = np.array(self.supportLabels)
        self.b1 = self.supportLabels[0] - np.matmul(self.supportVectors[0].T, self.w)
        
        b = []
        for i in tqdm(range(len(self.supportLabels))):
            b_i = self.supportLabels[i] - np.matmul(self.supportVectors[i].T, self.w)
            b.append( b_i )
        
        self.b2 = sum(b)/len(b)
        
        self.intercept = self.b2
        
    '''predicted classes''' 
    def predict(self, X):
        return 2*(np.matmul(X, self.w) + self.intercept > 0) - 1
    
    '''evaluate the performance of the model'''
    def evaluate(self,y,y_hat):
        y = ( y == 1)
        y_hat = (y_hat == 1)
        
        accuracy = (y == y_hat).sum()/ y.size
        precision =(y & y_hat).sum() / y_hat.sum()
        recall = (y & y_hat).sum()/ y.sum()
        f1score = 2 * (precision * recall) / (precision + recall)
        
        return accuracy,precision,recall,f1score
        
    '''run the model'''
    def runModel(self):
        self.X_train,self.X_test,self.y_train,self.y_test=train_test_split(self.X,
                                                                           self.y,
                                                                           stratify=self.y,
                                                                           test_size=0.2,
                                                                           random_state=1)
        self.fit(self.X_train,self.y_train)
        
        y_hat_train = self.predict(self.X_train)
        accuracy,precision,recall,f1score=self.evaluate(self.y_train,y_hat_train)
        print("=" * 50)
        print('Metrics for y_train:')
        print("=" * 50)
        metrics_labels = ['Accuracy', 'Precision', 'Recall', 'F1 score']
        metrics_values = [[accuracy, precision, recall, f1score]]
        print(pd.DataFrame(metrics_values, columns = metrics_labels, index = ['value']))
        print(f"\nConfusion Matrix:\n{pd.DataFrame(confusion_matrix(self.y_train, y_hat_train))}\n")
        
        y_hat_test = self.predict(self.X_test)
        accuracy,precision,recall,f1score=self.evaluate(self.y_test,y_hat_test)
        print("=" * 50)
        print('Metrics for y_test:')
        print("=" * 50)
        metrics_labels = ['Accuracy', 'Precision', 'Recall', 'F1 score']
        metrics_values = [[accuracy, precision, recall, f1score]]
        print(pd.DataFrame(metrics_values, columns = metrics_labels, index = ['value']))
        print(f"\nConfusion Matrix:\n{pd.DataFrame(confusion_matrix(self.y_test, y_hat_test))}\n")
        self.unseen_test(self.X_test_unseen,self.y_test_unseen)

    
    def unseen_test(self,X_test,y_test):
        y_hat_test = self.predict(X_test)
        accuracy,precision,recall,f1score=self.evaluate(y_test,y_hat_test)
        metrics_labels = ['Accuracy', 'Precision', 'Recall', 'F1 score']
        metrics_values = [[accuracy, precision, recall, f1score]]
        print("=" * 50)
        print('Metrics for UNSEEN y_test:')
        print("=" * 50)
        print(pd.DataFrame(metrics_values, columns = metrics_labels, index = ['value']))
        print(f"\nConfusion Matrix:\n{pd.DataFrame(confusion_matrix(y_test, y_hat_test))}\n") 
```
